## Call the CAD api for each of the objects

```python
query_string = 'https://ssd-api.jpl.nasa.gov/cad.api?date-min=1900-01-01&date-max=2100-01-01&des='

final_list_of_objects = []

for descriptor in sentry_list:
    full_query_string = query_string + descriptor
#     print(full_query_string)

    #get a response
    response1 = requests.get(full_query_string)

    #covnert response into json
    object_json = response1.json()
    
    if int(object_json['count']) >= 5:
        final_list_of_objects.append(object_json)
```

```python
columns = final_list_of_objects[0]['fields']
final_df = pd.DataFrame(columns = columns)
for dataframe in final_list_of_objects:
    final_object_df = pd.DataFrame(dataframe['data'], columns = dataframe['fields'])
    final_df = final_df.append(final_object_df, ignore_index=True)
```

```python
object_list = final_df['des'].unique()
object_list
```

## Call different Sentry API to get other data

```python
des_list = [descriptor.replace(' ', '%20') for descriptor in merged_df['des'].unique()]
```

```python
query_string1 = 'https://ssd-api.jpl.nasa.gov/sentry.api?des='

final_sentry_list_of_objects = []

for descriptor in des_list:
    full_query_string1 = query_string1 + descriptor
#     print(full_query_string1)

    #get a response
    response2 = requests.get(full_query_string1)

    #covnert response into json
    object_json2 = response2.json()
    
    final_sentry_list_of_objects.append(object_json2)
```

```python
columns_summary = final_sentry_list_of_objects[0]['summary'].keys()
columns_data = final_sentry_list_of_objects[0]['data'][0].keys()
summary_df = pd.DataFrame(columns = columns_summary)
data_df = pd.DataFrame(columns = columns_data)
for dataframe in final_sentry_list_of_objects:
    print(dataframe)
    final_des_df = pd.DataFrame(dataframe['data'], columns = columns_data)
    final_des_df['des'] = dataframe['summary']['des']
    print(dataframe['summary']['des'])
    data_df = data_df.append(final_des_df, ignore_index=True)
    summary_df = summary_df.append(pd.DataFrame(dataframe['summary'], columns = columns_summary, index = [0]))
```
