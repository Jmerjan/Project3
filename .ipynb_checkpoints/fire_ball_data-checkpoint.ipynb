```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
from pprint import pprint
import json
```

```python
# Create URL string
url = "https://ssd-api.jpl.nasa.gov/fireball.api?"
query = "date-min=2021-01-01&req-alt=true"
response = requests.get(url + query)
```

```python
#Get response
data = response.json()
pprint(data)
```

```python
# Assign data a variable
fire_ball = data['data']
fire_ball
```

```python
# Create pandas dataframe
fire_ball_data = pd.DataFrame(fire_ball)
fire_ball_data.head()
```

```python
# rename columns
fire_ball_data.rename(columns = {0:'Date', 
                                1: 'Energy', 
                                2: 'Impact Energy', 
                                3: 'Latitude', 
                                4: 'Lat Direction', 
                                5: 'Longitude', 
                                6: 'Lon Direction', 
                                7: 'Altitude (km)', 
                                8: 'Velocity (km/s)'
                                }, inplace = True)
fire_ball_data.head()
```

```python
# insert placeholder column for 'time'
fire_ball_data['Time'] = ""
fire_ball_data.head()
```

```python
#reorder columns
fire_ball_data = fire_ball_data.reindex(columns = ['Date', 
                                                   'Time', 
                                                   'Energy', 
                                                   'Impact Energy', 
                                                   'Latitude', 
                                                   'Lat Direction', 
                                                   'Longitude', 
                                                   'Lon Direction', 
                                                   'Altitude (km)', 
                                                   'Velocity (km/s)'
                                                  ])
fire_ball_data.head()
```

```python
# split date and time into seperate columns
fire_ball_data['Date'] = pd.to_datetime(fire_ball_data['Date'])

fire_ball_data = fire_ball_data.assign(Date=fire_ball_data.Date.dt.date, Time=fire_ball_data.Date.dt.time)
```
