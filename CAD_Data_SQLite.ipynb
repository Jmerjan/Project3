```python
for index, descriptor in enumerate(sentry_list):
    if ' ' in descriptor:
        string_array = descriptor.split(' ')
        full_string = string_array[0] + '%20' + string_array[1]
        sentry_list[index] = full_string

print(sentry_list)
```

## Call the CAD api for each of the objects

```python
query_string = 'https://ssd-api.jpl.nasa.gov/cad.api?date-min=1900-01-01&date-max=2100-01-01&des='

final_list_of_objects = []

for descriptor in sentry_list:
    full_query_string = query_string + descriptor
#     print(full_query_string)

    #get a response
    response1 = requests.get(full_query_string)

    #covnert response into json
    object_json = response1.json()
    
    if int(object_json['count']) >= 5:
        final_list_of_objects.append(object_json)
```

```python
# Define our Summary table
class Summary(Base):
    __tablename__ = 'summary'
    energy = Column(Float)
    darc = Column(String)
    ip = Column(Float)
    h = Column(Float)
    nobs = Column(Integer)
    mass = Column(Float)
    v_inf = Column(Float)
    first_obs = Column(String)
    method = Column(String)
    pdate = Column(String)
    ndop = Column(Integer)
    cdate = Column(String)
    ps_cum = Column(Float)
    diameter = Column(Float)
    ndel = Column(Integer)
    v_imp = Column(Float)
    ps_max = Column(Float)
    last_obs = Column(String)
    fullname = Column(String)
    n_imp = Column(Integer)
    ts_max = Column(Float)
    nsat = Column(Integer)
    des = Column(String, primary_key = True)
```

```python
# Define our Sentry table
class Sentry(Base):
    __tablename__ = 'sentry'
    id = Column(Integer, primary_key = True)
    energy = Column(Float)
    ip = Column(Float)
    ts = Column(Float)
    date = Column(String)
    sigma_vi = Column(Float)
    ps = Column(Float)
    des = Column(String)
```

```python
# Define our Cad table
class Cad(Base):
    __tablename__ = 'cad'
    des = Column(String)
    orbit_id = Column(Integer)
    jd = Column(Float)
    cd = Column(String)
    dist = Column(Float)
    dist_min = Column(Float)
    dist_max = Column(Float)
    v_rel = Column(Float)
    v_inf = Column(Float)
    t_sigma_f = Column(String)
    h = Column(Float)
    id = Column(Integer, primary_key = True)
```

```python
engine = create_engine("sqlite:///cad.sqlite")
conn = engine.connect()
Base.metadata.create_all(engine)
```
